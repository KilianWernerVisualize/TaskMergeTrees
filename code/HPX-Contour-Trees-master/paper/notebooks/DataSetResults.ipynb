```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PUMA = [49, 11.83, 5.4, 5.1]
FTM = [8.05, 4.53, 3.6, 0]
CF = [13.7, 8.45, 11.2, 9]
GPUMA = [2.47, 0.77, 1.72, 45]
VTKRG = [0, 0, 0, 0]

ind = np.arange(len(PUMA))  # the x locations for the groups
width = 0.7  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/8, [x / y for x,y in zip(CF,CF)], width/8,
                color='SkyBlue', label='TTK Contour Forests')
rects2 = ax.bar(ind, [x / y for x,y in zip(PUMA,CF)], width/8,
                color='IndianRed', label='PUMA')
rects3 = ax.bar(ind + width/8, [x / y for x,y in zip(FTM,CF)], width/8,
                color='Green', label='TTK FTM')
rects4 = ax.bar(ind + 2*width/8, [x / y for x,y in zip(GPUMA,CF)], width/8,
                color='Orange', label='GPUMA')



# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Runtime of TTK Contour Forests')
ax.set_ybound(0, 1.5)
ax.set_title('Runtime by dataset and algorithm')
ax.set_xticks(ind)
ax.set_xticklabels(('Jet', 'Meteor', 'ctBones', 'Dragon'))
ax.legend(bbox_to_anchor=(1.005, 1.05))


plt.savefig('runtime2.pdf', bbox_inches='tight')
```
